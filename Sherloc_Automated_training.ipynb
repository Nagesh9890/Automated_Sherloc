```python
# Pickle file prediction
import numpy as np
import pandas as pd
import pickle
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_extraction.text import TfidfVectorizer
import re

def custom_tokenizer(text):
    pattern = re.compile(r'[a-zA-Z]+\d+')
    return pattern.findall(text)

def predict_categories_demo(input_data):
    # Load the saved models and vectorizers
    clf_cat1 = pickle.load(open("clf_cat1_demo.pkl", "rb"))
    clf_cat2 = pickle.load(open("clf_cat2_demo.pkl", "rb"))

    tfidf_source_name = pickle.load(open("tfidf_source_name.pkl", "rb"))
    tfidf_mode_name = pickle.load(open("tfidf_mode_name.pkl", "rb"))
    tfidf_benef_name_type = pickle.load(open("tfidf_benef_name_type.pkl", "rb"))
    tfidf_nobook_txn_text_vpa = pickle.load(open("tfidf_nobook_txn_text_vpa.pkl", "rb"))

    # Transform input data using the TFIDF vectorizers
    source_name = tfidf_source_name.transform([input_data['source']])
    mode_name = tfidf_mode_name.transform([input_data['mode']])
    benef_name_type = tfidf_benef_name_type.transform([input_data['benef_name']])
    nobook_txn_text_vpa = tfidf_nobook_txn_text_vpa.transform([input_data['nobook_txn_text']])

    tfidf_matrix = pd.concat([pd.DataFrame(source_name.toarray()),
                              pd.DataFrame(mode_name.toarray()),
                              pd.DataFrame(benef_name_type.toarray()),
                              pd.DataFrame(nobook_txn_text_vpa.toarray())], axis=1)

    # Predict
    prediction_cat1 = clf_cat1.predict(tfidf_matrix)
    prediction_cat2 = clf_cat2.predict(tfidf_matrix)

    return prediction_cat1[0], prediction_cat2[0]

# Example usage
input_data_demo = {
    'source': 'XYZ Source',
    'mode': 'Online',
    'benef_name': 'Beneficiary Name',
    'nobook_txn_text': 'txn123text'
}

category1_demo, category2_demo = predict_categories_demo(input_data_demo)
print(f"Predicted Category Level 1: {category1_demo}")
print(f"Predicted Category Level 2: {category2_demo}")
```
